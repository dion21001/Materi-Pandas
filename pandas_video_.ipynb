# **Video 19:Formatting tampilan DataFrame**

```python
n_rows =5
n_cols =2
cols =["omset","operasional"]

df =pd.DataFrame(np.random.randint(1,20,size =(n_rows,n_cols)),
                 columns =cols)

df
```

```python
df["omset"] =df["omset"]*100_000
df["operasional"] =df["operasional"]*100_000
df
```

```python
df.index =pd.util.testing.makeDateIndex(n_rows,freq ="D")
df =df.reset_index()
df =df.rename(columns ={"index":"tanggal"})
df
```

**Melakukan formatting tampilan dataframe**

```python
formatku ={"tanggal":"{:%d/%m/%y}",
           "oprasional":"Rp{:.2f}",
           "omset":"Rp{:.2f}"}

laporan =df.style.format(formatku)
laporan
```

```python
type(laporan)
```

```python
laporan.hide_index()
```

```python
laporan.set_caption("Data Omset dan Operasional")
```

```python
laporan.highlight_min("omset",color ="pink")
laporan.highlight_max("omset",color ="lightgreen")

laporan.highlight_min("operasional",color ="lightblue")
laporan.highlight_max("operasional",color ="grey")
```

**Video 33 : Nilai kumulatif pada pandas data Frame**

**Persiapan data**

```python
d ={"Pemain":["budi","joni","iwan","budi","budi","iwan","asep","joni"],
    "goal":[2,1,3,1,1,2,2,3]}

df =pd.DataFrame(d)
df
```

**Menampilkan nilai kumulatif (cummulative)**

```python
df["goal"].cumsum().to_frame() # menurut index, cumcum  =cumulative sum
```

```python
df["jumlah_goal_kumulatif"] =df["goal"].cumsum()
df
```

```python
df["jumlah_goal_kumulatif_tiap_pemain"] =df.groupby("Pemain")["goal"].cumsum()

df
```

```python
df["cunmax"]  =df["goal"].cummax()# pemeringkatan terhadap goal terhadap nilai indeks lihat cunmax baris ke 3 yaitu 3 artinya sampai baris ke tiga itu nilai goal tertinggi adalah 3 begitu seterusnya
df
```

```python
df["cunmax"]  =df["goal"].cummin()
df
```

```python
df["cunmax"]  =df["goal"].cumprod()# baris ke 3 , 6 =3 dikali 1 dikali 2 pada goal,begitu seterusnya
df
```
